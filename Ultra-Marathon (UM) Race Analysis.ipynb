### DATA CLEANING

 #### Filtering 50km and 50 Mile USA Races in 2020 

```python
# Checking for "50mi and 50km" in Event distance/length column



# Checking for "50km"

df[df["Event distance/length"] == "50km"]
```

```python
# Checking for "50mi"


df[df["Event distance/length"] == "50mi"]
```

```python
# Checking for both "50km and 50mi"


df[df["Event distance/length"].isin(["50km","50mi"])]
```

```python
# Checking "50km and 50mi" Event distance/length in event year "2020"


df[(df["Event distance/length"].isin(["50km","50mi"])) & (df["Year of event"] == 2020)]
```

```python
# Checking "USA" Event name



df[df["Event name"].str.split('(').str.get(1).str.split(')').str.get(0) == "USA"]
```

```python
# checking for "50km and 50mi" Event distance/length in event year "2020" in "USA"


df[(df["Event distance/length"].isin(["50km","50mi"])) & (df["Year of event"] == 2020) & (df["Event name"].str.split('(').str.get(1).str.split(')').str.get(0) == "USA")]
```

```python
# Filtering and Naming "50km" and "50 Mile" "USA" Races in "2020" as "df2"


df2 = df[(df["Event distance/length"].isin(["50km","50mi"])) & (df["Year of event"] == 2020) & (df["Event name"].str.split('(').str.get(1).str.split(')').str.get(0) == "USA")]
```

```python
df2.head(10)
```

```python
df2.shape
```

```python
# Removing "(USA)" from event name (all events in df2 took place in USA)


df2["Event name"] = df2["Event name"].str.split('(').str.get(0)
```

```python
df2.head()
```

```python
# Adding a new column ("athelete_age")


df2["athlete_age"] = 2020 - df2["Athlete year of birth"]
```

```python
# Removing "h" from athlete performance column


df2["Athlete performance"] = df2["Athlete performance"].str.split(' ').str.get(0)
```

```python
df2.head(5)
```

```python
# Dropping columns that are not needed: Athlete club, Athlete country, Athlete year of birth, Athlete age category


df2 = df2.drop(["Athlete club", "Athlete country", "Athlete year of birth", "Athlete age category"], axis = 1)
```

```python
df2.head()
```

```python
# Cleaning up null values


df2.isna().sum()
```

```python
# Checking for null values in "athlete_age" column


df2[df2["athlete_age"].isna()==1]
```

```python
# Dropping rows with null values (athlete_age)


df2 = df2.dropna()
```

```python
df2.shape
```

```python
# Checking for duplicates


df2[df2.duplicated() == True]
```

```python
# Resetting index


df2.reset_index(drop = True)
```

```python
# Changing data types: "athlete_age", "Athlete average speed"


df2.dtypes
```

```python
# Changing "athlete_age" data type to integer


df2["athlete_age"] = df2["athlete_age"].astype(int)
```

```python
# Changing "Athlete average speed" data type to float


df2["Athlete average speed"] = df2["Athlete average speed"].astype(float)
```

```python
df2.dtypes
```

```python
df2.head()
```

```python
# Renaming columns


#Year of event                  int64
#Event dates                   object
#Event name                    object
#Event distance/length         object
#Event number of finishers      int64
#Athlete performance           object
#Athlete gender                object
#Athlete average speed        float64
#Athlete ID                     int64
#athlete_age                    int32


df2 = df2.rename(columns = {"Year of event":"year",
                            "Event dates":"race_day",
                            "Event name":"race_name",
                            "Event distance/length":"race_length",
                            "Event number of finishers":"race_number_of_finishers",
                            "Athlete performance":"athlete_performance",
                            "Athlete gender":"athlete_gender",
                            "Athlete average speed":"athlete_average_speed",
                            "Athlete ID":"athlete_id"
                           })
```

```python
df2.head()
```

```python
# Reordering columns


df2 = df2[["race_day", "race_name", "race_length", "race_number_of_finishers", "athlete_id", "athlete_gender", "athlete_age", "athlete_performance", "athlete_average_speed"]]
```

### 9. Analyze how well athletes performed across different seasons of the year.

```python
# There are four seasons in USA

# Spring - From the 3rd to the 5th month of the year
# Summer - From the 6th month to the 8th month of the year
# Fall - From the 9th month to the 11th month of the year 
# Winter - From the 12th month to the 2nd month of the year 
```

```python
# Splitting month number from "race_day" column into a new column ("race_month")


df2["race_month"] = df2["race_day"].str.split('.').str.get(1).astype(int)
```

```python
df2.head(25)
```

```python
# Creating a new column that assigns each "race_month" to its corresponding season


df2["race_season"] = df2["race_month"].apply(lambda x: "Winter" if x > 11 else "Fall" if x > 8 else "Summer" if x > 5 else "Spring" if x > 2 else "Winter")
```

```python
df2.head(25)
```

```python
# Performances across various seasons


df2.groupby("race_season")["athlete_average_speed"].agg(["mean", "count"]).sort_values("mean", ascending = False)
```

```python
# Performances of 50 milers only across various seasons


df2.query('race_length == "50mi"').groupby("race_season")["athlete_average_speed"].agg(["mean", "count"]).sort_values("mean", ascending = False)
```

```python
# Performances of 50km racers only across various seasons


df2.query('race_length == "50km"').groupby("race_season")["athlete_average_speed"].agg(["mean", "count"]).sort_values("mean", ascending = False)
```
